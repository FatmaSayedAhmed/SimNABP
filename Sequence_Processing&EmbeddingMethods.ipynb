# Create tokenizer for 3Kmer AAs (NB_WORDS = 8002), 1- 8000 for 3kmer AAs , 0 for padding , and 8001 for unk


```python
from keras.preprocessing.text import Tokenizer
import itertools
def get_tokenizer():
    AA = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
    c = itertools.product(AA,AA,AA)
    res=[]
    for i in c:
        temp=i[0]+i[1]+i[2]
        res.append(temp)
    res=np.array(res)
    #print(len(res))
    NB_WORDS = 8002
    tokenizer = Tokenizer(num_words=NB_WORDS)
    tokenizer.fit_on_texts(res)
    unk_AA_index = tokenizer.word_index
    unk_AA_index['<pad>'] = 0
    unk_AA_index = tokenizer.word_index
    unk_AA_index['<unk>'] = 8001

    return tokenizer
```

# Method to spilt sequences to 3kmer tokens



```python
def sentence2word(str_set):
    word_seq=[]
    for sr in str_set:
        tmp=[]
        for i in range(len(sr)-2):
            if('B' in sr[i:i+3]):
                tmp.append('<unk>')
            elif('J' in sr[i:i+3]):
                tmp.append('<unk>')
            elif('O' in sr[i:i+3]):
                tmp.append('<unk>')
            elif('U'  in sr[i:i+3]):
                tmp.append('<unk>')
            elif('X' in sr[i:i+3]):
                tmp.append('<unk>')
            elif('Z' in sr[i:i+3]):
                tmp.append('<unk>')
            else:
                tmp.append(sr[i:i+3])
        word_seq.append(' '.join(tmp))
    return word_seq
```

# Method to convert 3kmer tokens to integer numbers according to the 3kmer tokenizer and apply pad_sequences

```python
from keras.utils import pad_sequences
def word2num(wordseq,tokenizer,MAX_LEN):
    sequences = tokenizer.texts_to_sequences(wordseq)
    numseq = pad_sequences(sequences, maxlen=MAX_LEN)
    return numseq, sequences
```

# Method to convert sequences to integer numbers

```python
def sentence2num(str_set,tokenizer,MAX_LEN):
    wordseq=sentence2word(str_set)
    numseq, sequences =word2num(wordseq,tokenizer,MAX_LEN)
    return numseq, wordseq, sequences
```

```python
def sentence2AA(str_set):
    AA_seq=[]
    for sr in str_set:
        tmp=[]
        for i in range(len(sr)):
          tmp.append(sr[i])
        AA_seq.append(' '.join(tmp))
    return AA_seq
```

# Method to convert 1kmer tokens to integer numbers according to the 1kmer tokenizer and apply pad_sequences

```python
from keras.utils import pad_sequences
def AA2num(AA_seq,tokenizer,MAX_LEN):
    sequences = tokenizer.texts_to_sequences(AA_seq)
    numseq = pad_sequences(sequences, maxlen=MAX_LEN)
    return numseq, sequences
```

# Method to convert sequences to integer numbers
